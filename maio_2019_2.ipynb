```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
df = pd.read_csv('metar_sbgr_maio_2019.csv',index_col=1,squeeze=True,parse_dates=True,)
df=df.drop(['station'], axis=1)
df.head()
```

```python
df.info()
```

```python
df.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
df[['tmpc']].plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Data', fontsize=20);
```

```python
tmpc = df[['tmpc']]
tmpc.rolling(12).mean().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
relh = df[['relh']]
relh.rolling(12).mean().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
vsby = df[['vsby']]
vsby.rolling(12).mean().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
dwpc = df[['dwpc']]
dwpc.rolling(12).mean().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
df_rm = pd.concat([tmpc.rolling(12).mean(), dwpc.rolling(12).mean()], axis=1)
df_rm.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
vsby.diff().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
tmpc.diff().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
df.tmpc.diff().plot(figsize=(20,10), linewidth=5, fontsize=20)
df.dwpc.diff().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```
